```python
import pandas as pd
from geopy.distance import geodesic
from tqdm import tqdm

# Load data from Excel files
df1 = pd.read_excel('Координаты островок - удаленные из кср.xlsx')
df2 = pd.read_excel('Координаты Суточно РУ.xlsx')

# Define a function to calculate geographical distance
def calculate_distance(row1, row2):
    return geodesic((row1['latitude'], row1['longitude']), (row2['latitude'], row2['longitude'])).meters

# Initialize an empty list to store results
results = []

# Iterate over each row in df1 with tqdm progress bar
for index1, row1 in tqdm(df1.iterrows(), total=len(df1), desc="Calculating distances"):
    distances = []
    # Iterate over each row in df2
    for index2, row2 in df2.iterrows():
        # Calculate distance between coordinates
        distance = calculate_distance(row1, row2)
        # Store distance and ID from df2
        distances.append({'ID': row2['ID'], 'longitude': row2['longitude'], 'latitude': row2['latitude'], 'distance': distance})
    # Sort distances and select the three shortest ones
    distances.sort(key=lambda x: x['distance'])
    shortest_distances = distances[:3]
    # Add results to the list
    result = [row1['ID'], row1['longitude'], row1['latitude']]
    for i in range(3):
        if i < len(shortest_distances):
            result.extend([shortest_distances[i]['ID'], shortest_distances[i]['longitude'], shortest_distances[i]['latitude'], shortest_distances[i]['distance']])
        else:
            result.extend([None, None, None, None])
    results.append(result)

# Create a DataFrame from the results list
columns = ['ID1', 'Long1', 'Lat1', 'ID2_1', 'Long2_1', 'Lat2_1', 'Shortest_distance_1', 'ID2_2', 'Long2_2', 'Lat2_2', 'Shortest_distance_2', 'ID2_3', 'Long2_3', 'Lat2_3', 'Shortest_distance_3']
result_df = pd.DataFrame(results, columns=columns)
result_df
```
